### Fractals with numpy

Sources:

* [Colorful Fractals](https://tomroelandts.com/articles/how-to-compute-colorful-fractals-using-numpy-and-matplotlib)
* [Fast Fractals with numpy](https://thesamovar.wordpress.com/2009/03/22/fast-fractals-with-python-and-numpy/)

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# max_iter = 1024
# center = -1.1195+0.2718j
# extent = 0.005+0.005j

max_iter = 256
width = 256
height = 256
center = -0.8+0.0j
extent = 3.0+3.0j
scale = max((extent / width).real, (extent / height).imag)

result = np.zeros((height, width), int)
for j in range(height):
    for i in range(width):
        c = center + (i - width // 2 + (j - height // 2)*1j) * scale
        z = 0
        for k in range(max_iter):
            z = z**2 + c
            # if (z * z.conjugate()).real > 4.0:
            if abs(z) > 2.0:
                break
        result[j, i] = k
```
